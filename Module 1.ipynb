# **Flights Section**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import flatten_json
import os
from glob import glob
from pandas.io.json import json_normalize
```

```python
PATH = r"C:\Users\Vedanth.LAPTOP-MHC2LDIT\ML Project\Original Data"
EXT = "*.csv"
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]

df_from_each_file = (pd.read_csv(f) for f in all_csv_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
airports = ['ATL','CLT','DEN','DFW','EWR','IAH','JFK','LAS','LAX','MCO','MIA','ORD','PHX','SEA','SFO']
```

```python
req_columns = ['FlightDate','Quarter','Year','Month','DayofMonth' ,'DepTime','DepDel15','CRSDepTime','DepDelayMinutes','OriginAirportID','DestAirportID','ArrTime','CRSArrTime','ArrDel15','ArrDelayMinutes','Origin','Dest']

unreq_columns = []

for i in concatenated_df.columns :  
    if i not in req_columns:   
        unreq_columns.append(i)
```

```python
concatenated_df.drop(unreq_columns,axis=1,inplace=True).drop_duplicates()
```

```python
flights_data = concatenated_df[(concatenated_df['Origin'].isin(airports)) & (concatenated_df['Dest'].isin(airports))]
```

```python
flights_data.to_csv('flights_data.csv')
```

```python
flights_data = pd.read_csv('flights_data.csv')
```

```python
flights_data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
def coversion(time) :
    a = time %100
    if(a<30) :
        time = time - a
    elif(a>=30) :
        a = time + (100 - a)
    if(time == 2400):
        time = 0
    return time
```

```python
flights_data['time'] = flights_data['CRSArrTime'].apply(conversion).astype(str)
```

```python
flights_data['primary'] = flights_data['Dest']+'|'+ flights_data['FlightDate']+'|' + flights_data['time']
```

# **Weather Section**

```python
weather = ['windspeedKmph','winddirDegree','weatherCode','precipMM','visibility','pressure','cloudcover','DewPointF','WindGustKmph','tempF','WindChillF','humidity','date','time', 'airports']
weather_df = pd.DataFrame()
```

```python
PATH = r"C:\Users\Vedanth.LAPTOP-MHC2LDIT\ML Project\Original Data\weather-report"
EXT = "*.json"
all_json_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]
```

```python
req_files = []
for file in all_json_files:
    if '2016' in file:
        req_files.append(file)
    elif '2017' in file:
        req_files.append(file)
```

```python
for filename in req_files:
    with open(filename) as f:
        data = json.load(f) 
    var1 = json_normalize(data['data']['weather'],record_path='hourly',meta='date')
    var1['airports'] = data['data']['request'][0]['query'][0:3]
    var1 = var1.filter(weather)
    weather_df = pd.concat([weather_df,var1])
```

```python
weather_df['primary'] = weather_df['airports']+'|'+weather_df['date']+'|'+weather_df['time'] 
```

# **Merging Section**

```python
dataset = pd.merge(flights_data,weather_df,on='primary',how='inner')
```
